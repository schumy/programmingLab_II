Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.

Extra:

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
#prova
np.random.seed(0)
dist = sp.stats.norm(loc=0, scale=1)
sample = dist.rvs(size=10000) #random variates sample, genera l'array dalla distribuzione
plt.hist(sample, bins=50, density=True) #density normalizza l'y, (posso confrontare con la curva)
x = np.linspace(-5, 5)
plt.plot(x, dist.pdf(x)) #pdf calcola la densità in un valore specifico,
#in questo caso nei valori di x e lo fa sulla distribuzione dist (normale)
```

```python
# Interpolazione
x = np.linspace(0, 10, 10)
y = np.sin(x)

# Interpolazione lineare
f_interp = sp.interpolate.interp1d(x, y) #aggiungendo kind='cubic' l'interpolazione viene fatta su una curva e aumentare la densità rende la curva più morbida
x_dense = np.linspace(0, 10, 200)
y_interp = f_interp(x_dense)
```

```python
#1
#poi rivedi e spiega
dist = sp.stats.gamma(1)
sample = dist.rvs(size=1000)
plt.hist(sample, bins=50, density=True)
x = np.linspace(0,10, 100)
plt.plot(x, dist.pdf(x))
a = sp.stats.gamma.fit(sample) #shape, loc: da dove parte, scale: la deformazione
y = dist.cdf(x)
plt.plot(x, y)
legenda = ['Teorico', 'CDF', 'Campioni']
plt.legend(legenda)
varianza_campionaria = sample.var()
varianza_campionaria
```
